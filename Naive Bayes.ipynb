```python
import pandas as pd
```

```python
data=pd.read_csv('CKD.csv')
```

```python
data
```

```python
data=pd.get_dummies(data,dtype=int,drop_first=True)
```

```python
data
```

```python
data.columns
```

```python
inde=data[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes']]
inde
```

```python
dep=data[['classification_yes']]
dep
```

```python
data['classification_yes'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
input_train,input_test,output_train,output_test=train_test_split(inde,dep,test_size=0.3,random_state=0)
```

```python
from sklearn.naive_bayes import GaussianNB
classifier=GaussianNB()
classifier.fit(input_train,output_train)
```

```python
predict=classifier.predict(input_test)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(output_test,predict)
print(cm)
```

```python
from sklearn.metrics import classification_report
clf=classification_report(output_test,predict)
print(clf)
```

```python
from sklearn.metrics import f1_score
f1=f1_score(output_test,predict,average='weighted')
f1
```
