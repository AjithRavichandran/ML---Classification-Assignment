```python
import pandas as pd
```

```python
data=pd.read_csv('CKD.csv')
```

```python
data
```

```python
data=pd.get_dummies(data,dtype=int,drop_first=True)
```

```python
data
```

```python
data.columns
```

```python
inde=data[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes']]
inde
```

```python
dep=data[['classification_yes']]
dep
```

```python
dep['classification_yes'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
input_train,input_test,output_train,output_test=train_test_split(inde,dep,test_size=0.3,random_state=0)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
param_grid={'criterion':['gini','entropy','log_loss'],
             'splitter':['best','random'],
         'max_features':['sqrt','log2']}
grid=GridSearchCV(DecisionTreeClassifier(),param_grid,refit=True,verbose=3,n_jobs=-1,scoring='f1_weighted')
```

```python
grid.fit(input_train,output_train)
```

```python
predict=grid.predict(input_test)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(output_test,predict)
print(cm)
```

```python
from sklearn.metrics import classification_report
clf=classification_report(output_test,predict)
print(clf)
```

```python
from sklearn.metrics import f1_score
f1=f1_score(output_test,predict,average='weighted')
f1
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(output_test,grid.predict_proba(input_test)[:,1])
```

```python
cla=grid.cv_results_
```

```python
table=pd.DataFrame.from_dict(cla)
```
